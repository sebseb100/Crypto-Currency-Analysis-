```python
#PYTHON LIBRARIES
from pandas_datareader import data,wb
import numpy as np
import datetime
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#Can't use interactive plots for python
import plotly
import cufflinks as cf
cf.go_offline()
```

```python
#Start and End Times for Price Data

start = datetime.datetime(2016,6,1)

end = datetime.datetime(2019,7,1)
```

```python
#3 Crypto Currencies

bitcoin   =   data.DataReader('BTC-USD','yahoo',start,end)

litecoin  =   data.DataReader('LTC-USD','yahoo',start,end)

ethereum  =   data.DataReader('ETH-USD','yahoo',start,end)
```

```python
#For later use lets combine all of the currency data into a sinlge dataframe

names     = ['Bitcoin','Ethereum','Litecoin']

all_coins = pd.concat([bitcoin,ethereum,litecoin],axis=1,keys=names)

all_coins.columns.names=['Currency','Price Data']

all_coins.head()
```

```python
#Correlation Plots

sns.jointplot(x= bitcoin['Close'].pct_change(),
                              
              y= litecoin['Close'].pct_change(),s=10,color='black')
```

```python
sns.jointplot(x= bitcoin['Close'].pct_change(),
                              
              y= ethereum['Close'].pct_change(),s=10,color='black')

#this shows the price distribution of the price percent change of bitcoin and ethereum mostly ranging from 10% 
```

```python
#Now lets analyze correlation using heatmaps 
plt.figure(figsize=(10,10))
plt.title('Correlation of BTC,LTC, and ETH')
okay = all_coins.xs(key='Close',axis=1,level='Price Data').corr()
sns.heatmap(okay,annot=True,cmap='Purples') #annot = annotation

#0.8 is said to be a very strong correlation, and the lowest correlation is 0.88 on this heatmap
```

```python
#Correlation Heatmap 

sns.clustermap(okay,annot=True,cmap='Reds')
```
